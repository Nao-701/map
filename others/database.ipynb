```python
import geopandas as gpd
import pandas as pd

data = gpd.read_file("shapefile/1741市区町村.shp")  
data = data.drop(["KEN", "SICHO", "SIKUCHOSON", "P_NUM", "H_NUM", "GIS用コ", "GIS用_1", 
"市区町","輸送用","非エネ", "最終エ", "エネル", "再エ_1", "Shape_Leng", "Shape_Area", "Ind_rate", "Ind_rate1",
"太陽光_","陸上風","中小水","地熱","洋上風","木質バ","再エネ","産業","業務","運輸","熱","電力","CITY_ENG","都道_1","都道_12","家庭"
], axis = 1)

csv = pd.read_csv("data.csv", encoding="shift-jis")  

df = pd.concat([data, csv],  axis=1)
print(sum(df["id"]-df["都道府"]))
df = df.drop(["id","都道府"], axis = 1)

df.to_file(r"data.geojson")
```
